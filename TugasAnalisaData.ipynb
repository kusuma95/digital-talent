### [1] Deskripsi Data

Dataset ini berisi data mengenai jumlah anak dengan status kesejahteraan 30% yang bersekolah berdasarkan tingkat pendidikan dan jenis kelamin Tahun 2012

```python
import pandas as pd
import matplotlib\

# Menampilkan data dalam bentuk tabel
data_pendidikan = pd.read_csv('processed-jumlah-partisipasi-sekolah-jenis-kelamin-30persen-termiskin-indonesia-perjuli2012.csv', delimiter = ',', encoding = "ISO-8859-1")
data_pendidikan
```

```python
# Menampilkan kolom yang terdapat pada tabel
data_pendidikan.columns
```

```python
# Menghitung jumlah data pada tabel
data_pendidikan.count()
```

```python
# Mengetahui discribsi statistik sederhana dari tabel status kesejahteraan
round(data_pendidikan.describe()['jumlah_individu'],2)
```

### [2] Statistik tentang data

```python
# Menampilkan data jumlah individu berdasarkan tingkat pendidikannya
data_pendidikan.groupby('tingkat_pendidikan').sum()['jumlah_individu']
```

### [3] Visualisasi Sederhana

```python
# Visualisai untuk menuntujukkan perbandingan jumlah individu Laki-laki dan Perempuan
ax = data_pendidikan.groupby('jenis_kelamin').sum()['jumlah_individu']
ax = ax.plot(kind='pie', title ="Perbandingan jumlah individu berdasarkan jenis kelamin", legend=False, figsize=(6,6), fontsize=10)
```

```python
# Visualisasiuntuk menunjukkan jumlah individu status kesejahteraan di setiap provinsi
ax = data_pendidikan.groupby('kode_provinsi').sum()['jumlah_individu']
ax = ax.plot(kind='bar', title ="Jumlah Individu setiap Provinsi", legend=False, fontsize=10, figsize=(15,7))
ax.set_xlabel("Kode Provinsi", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```
