# [-] Crawlink berita dari KOMPAS (kompas.com)

```python
# Import library yang digunakan
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import requests,re
```

```python
# Fungsi untuk mendapatkan semua link dari portal kompas.com
def getLink(link):
    req = Request(link)
    html_page = urlopen(req)
    soup = BeautifulSoup(html_page, "lxml")
    links = soup.find_all('a')
    for link in links:
        print('%s: %s'%(link.text.strip(),link["href"]))
```

```python
# Fungsi untuk mendapatkan isi berita dari salah satu link artikel di kompas.com
def crawLink(link):
    berita = []
    request = requests.get(link)
    html_page = request.content
    soup = BeautifulSoup(html_page, "lxml")
    judul = soup.find('h1',{'class':'read__title'})
    content = soup.find('div',{'class':'read__content'})
    isi = content.findAll('p')
    for div in isi:
        txt = div.text
        berita.append(txt)
    text_berita = ' '.join(berita)
    return[judul.text,text_berita]
```

```python
# Hasil link dari kompas.com
link_portal = 'https://www.kompas.com/'
result_link = getLink(link_portal)
print(result_link)
```

```python
# Hasil isi berita dari salah satu link artikel di kompas.com
link_kompas = 'https://regional.kompas.com/read/2018/10/29/20540851/jarak-pandang-terhalang-penyelaman-di-lokasi-jatuhnya-lion-air-jt-610'
result_berita = crawLink(link_kompas)
print('Judul Berita: ', result_berita[0])
print('Isi Berita: ', result_berita[1])
```
